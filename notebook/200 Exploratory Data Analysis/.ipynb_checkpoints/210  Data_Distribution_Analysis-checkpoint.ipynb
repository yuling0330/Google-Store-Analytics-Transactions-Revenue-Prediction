```python
x=transaction_df.drop(['id','revenue'],axis=1)
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
model=TSNE(learning_rate=100)
Tsne_transformed=model.fit_transform(x)
```

```python
xs =Tsne_transformed[:,0]
ys=Tsne_transformed[:,1]
plt.figure(figsize=(15, 10))
a = transaction_df['id'].values
label = a.tolist()
# Scatter plot
plt.scatter(xs, ys, alpha=0.5)

# # Annotate the points
# for x, y, flight in zip(xs, ys, label):
#     plt.annotate(flight,(x, y), fontsize=10, alpha=0.75)
# #plt.figure(figsize=(18,20))
plt.show()
```
